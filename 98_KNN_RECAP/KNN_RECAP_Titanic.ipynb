# K En Yakın Komşu (KNN - K-Nearest Neighbors)

**Teori**

KNN, gözetimli öğrenmenin basit ve etkili bir sınıflandırma yöntemidir. Yeni bir veri noktasını sınıflandırırken, eğitim setindeki "K" en yakın veri noktasına bakar ve bu "K" veri noktasının en yaygın sınıfını bu yeni veri noktası için tahmin olarak kullanır.

**Matematiksel Yönü**

KNN'nin temelinde uzaklık kavramı yatar. İki veri noktası arasındaki uzaklık, bu iki nokta arasındaki "mesafeye" dayanarak hesaplanır.

**Mesafe Ölçme Yöntemleri**

- **Öklidyen Mesafesi (Euclidean Distance):** İki nokta arasındaki en doğrudan uzaklık olarak tanımlanır.

\[
d(P, Q) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
\]

- **Manhattan Mesafesi:** İki nokta arasındaki ızgara üzerindeki minimum uzaklıktır.

\[
d(P, Q) = |x_2 - x_1| + |y_2 - y_1|
\]

**En İyi K Değerini Bulma**

K değeri, algoritmanın performansını doğrudan etkileyen kritik bir hiperparametredir.

- **Çapraz Doğrulama (Cross-Validation):** Farklı k değerleri için çapraz doğrulama yaparak en iyi sonucu veren k değerini seçebiliriz.
  
- **Elbow Yöntemi:** Hata oranını farklı k değerleri için hesaplayarak en düşük hata oranını veren k değerini seçebiliriz.

**Ölçeklendirme (Scaling)**

KNN, mesafe temelli olduğundan, özelliklerin değer aralıklarına son derece duyarlıdır.

- **Özelliklerin Ölçeği:** Eğer bir özellik diğerlerine göre çok daha büyük değerlere sahipse, KNN bu özelliğin değerlerini diğerlerinden daha baskın olarak algılar.

- **Özelliklerin Kalitesi:** İlgisiz veya gürültülü özelliklerin varlığı, KNN'nin performansını olumsuz etkileyebilir. Özellik seçimi, ilgisiz veya gürültülü özellikleri belirlemek için kritik öneme sahiptir.

**Ölçeklendirme Seçimi**

- **Dağılım ve Aykırı Değerler:** Eğer veri normal bir dağılıma sahipse, Z-Skor Ölçeklendirme (StandardScaler) uygun olabilir. Fakat, dengesiz dağılım veya fazla aykırı değer varsa, Min-Max (MinMaxScaler) Ölçeklendirme daha uygun olabilir.

- **Kategorik Veriler:** Kategorik özelliklerde, Min-Max Ölçeklendirme kullanmak anlamlı olmayabilir.

- **Model Bağımlılığı:** KNN gibi mesafe hesaplamalarına dayanan bir model, özelliklerin ölçeklendirilmesinden önemli ölçüde etkilenebilir.

**Son Notlar**

- **Veri Yoğunluğu:** Büyük veri setleri için, KNN'nin hesaplama yoğunluğu ideal değildir.
- **Ağırlıklandırma:** Tüm komşuların eşit olarak değerlendirilmediği durumlar için ağırlıklandırma kullanabiliriz.

---


Euclidean (Öklidyen) ve Manhattan mesafe ölçüm yöntemleri, KNN (K-Nearest Neighbors) algoritmasında uzaklık hesaplamalarında sıkça kullanılır. Her iki yöntemin de farklı avantajları ve kullanım senaryoları vardır. İşte bu iki yöntemi mukayese eden bir açıklama:

**1. Euclidean (Öklidyen) Mesafe:**

Bu yöntem, iki nokta arasındaki en kısa doğru çizgi uzunluğunu temsil eder. Euclidean mesafe, iki noktanın koordinatları arasındaki doğrudan uzaklığı hesaplar. Özellikle eğer veri uzayı homojen (her boyutun aynı ölçekte olduğu) ve dağılım normal ise, Öklidyen mesafe iyi bir seçim olabilir.

**Avantajları:**
- Hesaplaması basittir.
- Homojen veri uzaylarında etkili çalışır.
- 2 veya 3 boyutlu uzaylarda anlaşılırdır.

**Dezavantajları:**
- Uzaklık hesaplaması, bazen çok büyük veya çok küçük değerler verebilir.
- Çok yüksek boyutlu uzaylarda aşırı uyum (overfitting) riski vardır.
- Outlier'lardan (aykırı değerlere) olumsuz etkilenir

**2. Manhattan Mesafe:**

Bu yöntem, iki nokta arasındaki yolu temsil eder. Bir bakış açısına göre, Manhattan mesafesi, bir şehir bloğundaki iki nokta arasındaki yolu düşünmek gibidir. Bu nedenle "şehir bloğu mesafesi" olarak da adlandırılır. Eğer veri uzayı heterojen (farklı ölçeklerde boyutlar içeriyor) veya dağılım normalden uzaksa, Manhattan mesafesi kullanmak daha uygun olabilir.

**Avantajları:**
- Uzaklık hesaplaması daha düzgün ve outlier'lara karşı daha dirençlidir.
- Farklı ölçeklerdeki boyutlarla iyi çalışır.
- Yüksek boyutlu uzaylarda daha etkili olabilir.

**Dezavantajları:**
- Hesaplaması daha karmaşıktır, özellikle yüksek boyutlu uzaylarda.
- 2 veya 3 boyutlu uzaylarda Öklidyen mesafe kadar anlaşılır değildir.

Hangi yöntemi kullanmanın daha iyi olduğu, veri setinin özelliklerine ve problem bağlamına bağlıdır. Öklidyen mesafe, genellikle homojen veri uzayları için tercih edilirken, Manhattan mesafesi heterojen veri uzayları veya veri setinin dağılımı farklı olduğunda daha uygun olabilir. Ayrıca, hangi mesafe ölçüm yönteminin daha iyi performans gösterdiğini belirlemek için çapraz doğrulama gibi yöntemler kullanılabilir.

## Model Performance on Classification Tasks

```python
def eval_metric(model, X_train, y_train, X_test, y_test):

    """ to get the metrics for the model """

    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
    
# hem train hem de test seti scorlarını gözlemleyebilmek için eval_metric fornksiyonumuzu tanımlıyoruz.
```

## Elbow Method for Choosing Reasonable K Values

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
test_error_rates = []
train_error_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train) 
   
    y_pred_test = knn_model.predict(X_test_scaled)
    y_pred_train = knn_model.predict(X_train_scaled)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
    
    train_error = 1 - accuracy_score(y_train,y_pred_train)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));

```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));
```

# SMOTE for KNN

Dengesiz (unbalanced) veri setleri, sınıflandırma problemlerinde sıkça karşılaşılan bir durumdur. Özellikle nadir olayların tespit edildiği durumlarda (örn. dolandırıcılık tespiti, hastalık teşhisi) bu tür veri setleriyle karşılaşılır. Dengesiz veri setleriyle çalışırken modelin azınlık sınıfına yeterince öğrenemediği ve bu nedenle genellikle çoğunluk sınıfını tahmin etmeye eğilimli olduğu gözlemlenir.

Dengesiz veri setlerini ele almak için birçok yöntem bulunmaktadır:

1. **Örneklem Alma (Resampling) Yöntemleri**:
   - **Azınlık Sınıfını Artırma (Oversampling)**: Azınlık sınıfından ek örnekler üreterek bu sınıfın sayısını artırma yöntemidir. Bu, rastgele örnekleme veya SMOTE gibi algoritmalarla yapılabilir.
   - **Çoğunluk Sınıfını Azaltma (Undersampling)**: Çoğunluk sınıfından örneklerin rastgele kaldırılmasıyla bu sınıfın sayısını azaltma yöntemidir. 

2. **Sentetik Örnek Oluşturma**:
   - **SMOTE (Synthetic Minority Over-sampling Technique)**: Azınlık sınıfından sentetik örnekler üreterek azınlık sınıfının sayısını artırır.
   - **ADASYN (Adaptive Synthetic Sampling)**: SMOTE'a benzer bir tekniktir ancak sınıf dengesizliğini daha adaptif bir şekilde ele alır.

3. **Kümelenme Bazlı Örnekleme**: Veri seti kümelere ayrılır ve her kümeden örnekler alınarak dengesizlik azaltılır.

4. **Farklı Algoritmalar Kullanma**: Bazı algoritmalar dengesiz veri setleriyle daha iyi başa çıkabilir. Örneğin, `RandomForest` gibi bazı algoritmaların dengesiz veri setlerine karşı doğal bir direnci vardır.

5. **Cost-sensitive Öğrenme**: Modelin yanlış sınıflandırmalar için farklı maliyetler ödemesini sağlayarak dengesizliği ele alabilirsiniz.

6. **Farklı Değerlendirme Metrikleri Kullanma**: Dengesiz veri setlerinde, modelin performansını değerlendirmek için 'accuracy' yerine 'precision', 'recall', 'F1-score' gibi metrikleri kullanmak daha bilgilendirici olabilir.

7. **Topluluk Yöntemleri (Ensemble Methods)**: Dengesiz veri setleri için özelleştirilmiş topluluk yöntemleri kullanılabilir. Örneğin, BalancedRandomForestClassifier veya RUSBoost gibi yöntemler.

8. **Öğrenme oranını ayarlama**: Özellikle derin öğrenme modellerinde, dengesiz sınıflar için farklı öğrenme oranları atanabilir.

9. **Anomali Tespiti Yöntemleri**: Dengesiz veri setlerini bir anomali tespiti problemi olarak ele alabilirsiniz.

Bu yöntemlerin her biri, uygulandığı veri kümesine ve problemin doğasına bağlı olarak farklı sonuçlar verebilir. En iyi sonuçları elde etmek için bu yöntemleri deneyerek hangisinin en iyi çalıştığını belirlemek genellikle en iyi yaklaşımdır.

```python
# 1. Kütüphaneleri Al
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# 2. Veri Ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA Uygulama
pca = PCA(n_components=0.95)  # %95 varyansı koruyarak PCA uygula
X_pca = pca.fit_transform(X_scaled)

# 4. Veriyi Eğitim ve Test Setlerine Bölme
X_train, X_test, y_train, y_test = train_test_split(X_pca,
                                                    y, 
                                                    test_size=0.30, 
                                                    random_state=101)

# 5. SMOTE ile Oversampling
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
test_error_rates = []
train_error_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_smote,y_train_smote) 
   
    y_pred_test = knn_model.predict(X_test)
    y_pred_train = knn_model.predict(X_train_smote)
    
    test_error = 1 - accuracy_score(y_test, y_pred_test)
    test_error_rates.append(test_error)
    
    train_error = 1 - accuracy_score(y_train_smote,y_pred_train)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));
```
