# K En Yakn Komu (KNN - K-Nearest Neighbors)

**Teori**

KNN, gzetimli renmenin basit ve etkili bir snflandrma yntemidir. Yeni bir veri noktasn snflandrrken, eitim setindeki "K" en yakn veri noktasna bakar ve bu "K" veri noktasnn en yaygn snfn bu yeni veri noktas iin tahmin olarak kullanr.

**Matematiksel Yn**

KNN'nin temelinde uzaklk kavram yatar. ki veri noktas arasndaki uzaklk, bu iki nokta arasndaki "mesafeye" dayanarak hesaplanr.

**Mesafe lme Yntemleri**

- **klidyen Mesafesi (Euclidean Distance):** ki nokta arasndaki en dorudan uzaklk olarak tanmlanr.

\[
d(P, Q) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
\]

- **Manhattan Mesafesi:** ki nokta arasndaki zgara zerindeki minimum uzaklktr.

\[
d(P, Q) = |x_2 - x_1| + |y_2 - y_1|
\]

**En yi K Deerini Bulma**

K deeri, algoritmann performansn dorudan etkileyen kritik bir hiperparametredir.

- **apraz Dorulama (Cross-Validation):** Farkl k deerleri iin apraz dorulama yaparak en iyi sonucu veren k deerini seebiliriz.
  
- **Elbow Yntemi:** Hata orann farkl k deerleri iin hesaplayarak en dk hata orann veren k deerini seebiliriz.

**leklendirme (Scaling)**

KNN, mesafe temelli olduundan, zelliklerin deer aralklarna son derece duyarldr.

- **zelliklerin lei:** Eer bir zellik dierlerine gre ok daha byk deerlere sahipse, KNN bu zelliin deerlerini dierlerinden daha baskn olarak alglar.

- **zelliklerin Kalitesi:** lgisiz veya grltl zelliklerin varl, KNN'nin performansn olumsuz etkileyebilir. zellik seimi, ilgisiz veya grltl zellikleri belirlemek iin kritik neme sahiptir.

**leklendirme Seimi**

- **Dalm ve Aykr Deerler:** Eer veri normal bir dalma sahipse, Z-Skor leklendirme (StandardScaler) uygun olabilir. Fakat, dengesiz dalm veya fazla aykr deer varsa, Min-Max (MinMaxScaler) leklendirme daha uygun olabilir.

- **Kategorik Veriler:** Kategorik zelliklerde, Min-Max leklendirme kullanmak anlaml olmayabilir.

- **Model Bamll:** KNN gibi mesafe hesaplamalarna dayanan bir model, zelliklerin leklendirilmesinden nemli lde etkilenebilir.

**Son Notlar**

- **Veri Younluu:** Byk veri setleri iin, KNN'nin hesaplama younluu ideal deildir.
- **Arlklandrma:** Tm komularn eit olarak deerlendirilmedii durumlar iin arlklandrma kullanabiliriz.

---


Euclidean (klidyen) ve Manhattan mesafe lm yntemleri, KNN (K-Nearest Neighbors) algoritmasnda uzaklk hesaplamalarnda ska kullanlr. Her iki yntemin de farkl avantajlar ve kullanm senaryolar vardr. te bu iki yntemi mukayese eden bir aklama:

**1. Euclidean (klidyen) Mesafe:**

Bu yntem, iki nokta arasndaki en ksa doru izgi uzunluunu temsil eder. Euclidean mesafe, iki noktann koordinatlar arasndaki dorudan uzakl hesaplar. zellikle eer veri uzay homojen (her boyutun ayn lekte olduu) ve dalm normal ise, klidyen mesafe iyi bir seim olabilir.

**Avantajlar:**
- Hesaplamas basittir.
- Homojen veri uzaylarnda etkili alr.
- 2 veya 3 boyutlu uzaylarda anlalrdr.

**Dezavantajlar:**
- Uzaklk hesaplamas, bazen ok byk veya ok kk deerler verebilir.
- ok yksek boyutlu uzaylarda ar uyum (overfitting) riski vardr.
- Outlier'lardan (aykr deerlere) olumsuz etkilenir

**2. Manhattan Mesafe:**

Bu yntem, iki nokta arasndaki yolu temsil eder. Bir bak asna gre, Manhattan mesafesi, bir ehir bloundaki iki nokta arasndaki yolu dnmek gibidir. Bu nedenle "ehir blou mesafesi" olarak da adlandrlr. Eer veri uzay heterojen (farkl leklerde boyutlar ieriyor) veya dalm normalden uzaksa, Manhattan mesafesi kullanmak daha uygun olabilir.

**Avantajlar:**
- Uzaklk hesaplamas daha dzgn ve outlier'lara kar daha direnlidir.
- Farkl leklerdeki boyutlarla iyi alr.
- Yksek boyutlu uzaylarda daha etkili olabilir.

**Dezavantajlar:**
- Hesaplamas daha karmaktr, zellikle yksek boyutlu uzaylarda.
- 2 veya 3 boyutlu uzaylarda klidyen mesafe kadar anlalr deildir.

Hangi yntemi kullanmann daha iyi olduu, veri setinin zelliklerine ve problem balamna baldr. klidyen mesafe, genellikle homojen veri uzaylar iin tercih edilirken, Manhattan mesafesi heterojen veri uzaylar veya veri setinin dalm farkl olduunda daha uygun olabilir. Ayrca, hangi mesafe lm ynteminin daha iyi performans gsterdiini belirlemek iin apraz dorulama gibi yntemler kullanlabilir.

## Model Performance on Classification Tasks

```python
def eval_metric(model, X_train, y_train, X_test, y_test):

    """ to get the metrics for the model """

    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
    
# hem train hem de test seti scorlarn gzlemleyebilmek iin eval_metric fornksiyonumuzu tanmlyoruz.
```

## Elbow Method for Choosing Reasonable K Values

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
test_error_rates = []
train_error_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train) 
   
    y_pred_test = knn_model.predict(X_test_scaled)
    y_pred_train = knn_model.predict(X_train_scaled)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
    
    train_error = 1 - accuracy_score(y_train,y_pred_train)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));

```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));
```

# SMOTE for KNN

Dengesiz (unbalanced) veri setleri, snflandrma problemlerinde ska karlalan bir durumdur. zellikle nadir olaylarn tespit edildii durumlarda (rn. dolandrclk tespiti, hastalk tehisi) bu tr veri setleriyle karlalr. Dengesiz veri setleriyle alrken modelin aznlk snfna yeterince renemedii ve bu nedenle genellikle ounluk snfn tahmin etmeye eilimli olduu gzlemlenir.

Dengesiz veri setlerini ele almak iin birok yntem bulunmaktadr:

1. **rneklem Alma (Resampling) Yntemleri**:
   - **Aznlk Snfn Artrma (Oversampling)**: Aznlk snfndan ek rnekler reterek bu snfn saysn artrma yntemidir. Bu, rastgele rnekleme veya SMOTE gibi algoritmalarla yaplabilir.
   - **ounluk Snfn Azaltma (Undersampling)**: ounluk snfndan rneklerin rastgele kaldrlmasyla bu snfn saysn azaltma yntemidir. 

2. **Sentetik rnek Oluturma**:
   - **SMOTE (Synthetic Minority Over-sampling Technique)**: Aznlk snfndan sentetik rnekler reterek aznlk snfnn saysn artrr.
   - **ADASYN (Adaptive Synthetic Sampling)**: SMOTE'a benzer bir tekniktir ancak snf dengesizliini daha adaptif bir ekilde ele alr.

3. **Kmelenme Bazl rnekleme**: Veri seti kmelere ayrlr ve her kmeden rnekler alnarak dengesizlik azaltlr.

4. **Farkl Algoritmalar Kullanma**: Baz algoritmalar dengesiz veri setleriyle daha iyi baa kabilir. rnein, `RandomForest` gibi baz algoritmalarn dengesiz veri setlerine kar doal bir direnci vardr.

5. **Cost-sensitive renme**: Modelin yanl snflandrmalar iin farkl maliyetler demesini salayarak dengesizlii ele alabilirsiniz.

6. **Farkl Deerlendirme Metrikleri Kullanma**: Dengesiz veri setlerinde, modelin performansn deerlendirmek iin 'accuracy' yerine 'precision', 'recall', 'F1-score' gibi metrikleri kullanmak daha bilgilendirici olabilir.

7. **Topluluk Yntemleri (Ensemble Methods)**: Dengesiz veri setleri iin zelletirilmi topluluk yntemleri kullanlabilir. rnein, BalancedRandomForestClassifier veya RUSBoost gibi yntemler.

8. **renme orann ayarlama**: zellikle derin renme modellerinde, dengesiz snflar iin farkl renme oranlar atanabilir.

9. **Anomali Tespiti Yntemleri**: Dengesiz veri setlerini bir anomali tespiti problemi olarak ele alabilirsiniz.

Bu yntemlerin her biri, uyguland veri kmesine ve problemin doasna bal olarak farkl sonular verebilir. En iyi sonular elde etmek iin bu yntemleri deneyerek hangisinin en iyi altn belirlemek genellikle en iyi yaklamdr.

```python
# 1. Ktphaneleri Al
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# 2. Veri leklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA Uygulama
pca = PCA(n_components=0.95)  # %95 varyans koruyarak PCA uygula
X_pca = pca.fit_transform(X_scaled)

# 4. Veriyi Eitim ve Test Setlerine Blme
X_train, X_test, y_train, y_test = train_test_split(X_pca,
                                                    y, 
                                                    test_size=0.30, 
                                                    random_state=101)

# 5. SMOTE ile Oversampling
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
test_error_rates = []
train_error_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_smote,y_train_smote) 
   
    y_pred_test = knn_model.predict(X_test)
    y_pred_train = knn_model.predict(X_train_smote)
    
    test_error = 1 - accuracy_score(y_test, y_pred_test)
    test_error_rates.append(test_error)
    
    train_error = 1 - accuracy_score(y_train_smote,y_pred_train)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.xticks(range(0, 31, 1));
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')

plt.hlines(y=0.20, xmin = 0, xmax = 30, colors= 'r', linestyles="--") #Bunlar ekledim
plt.hlines(y=0.175, xmin = 0, xmax = 30, colors= 'r', linestyles="--")


plt.xticks(range(0, 31, 1));
```
